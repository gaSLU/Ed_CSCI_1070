23. (3 pts) Write a function that takes in any word and counts the number of a particular letter in it. Give the function a default value if no word is passed in. You should not call a python function within your function. 

```python
def count_letter(word="spider", letter="o"):
    count = 0
    for ch in word:
        if ch == letter:
            count += 1
    return count
```

Bonus (2 pts): Write pseudocode for the following: 

Given a string, return the first character that does not repeat.

Input: "locomotion"
Output: "l"

```python
def first_non_repeating_char(word):
    for ch in word:
        count = 0
        for c in word:
            if c == ch:
                count += 1
        if count == 1:
            return ch
```
